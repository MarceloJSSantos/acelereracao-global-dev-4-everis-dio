```
# cria um dataframe
rdd1 = [{"nome": "Marco","idade": 33,"status": 'true'},
{"nome": "Antonio","idade":33,"status": 'true'},
{"nome":"Pereira","idade":33,"status": 'true'},
{"nome":"Helena","idade":30,"status": 'true'},
{"nome":"Fernando","idade":35,"status": 'true'},
{"nome":"Carlos","idade":28,"status": 'true'},
{"nome":"Lisa","idade":26,"status": 'true'},
{"nome":"Candido","idade":75,"status": 'false'},
{"nome":"Vasco","idade":62,"status": 'true'}
]
dff1 = spark.createDataFrame(rdd1)
dff1.show()

# cria outro dataframe
rdd2 = [
{"nome":"Marco","PaisOrigem":"Brasil"},
{"nome":"Helena","PaisOrigem":"Brasil"},
{"nome":"Gabriel","PaisOrigem":"Brasil"},
{"nome":"Vasco","PaisOrigem":"Portugal"},
{"nome":"Medhi","PaisOrigem":"Marocco"}]

dff2 = spark.createDataFrame(rdd2)
dff2.show()
```

```
'''
join_type = "inner" (condição satisfeita nos 2 dfs, retorna as colunas dos 2 dfs)

É usada para juntar dois DataFrames / Conjuntos de dados em colunas de chave e onde as chaves não correspondem às linhas são eliminadas de ambos os conjuntos de dados

+------+-----+------+------+----------+
|  nome|idade|status|  nome|PaisOrigem|
+------+-----+------+------+----------+
| Vasco|   62|  true| Vasco|  Portugal|
| Marco|   33|  true| Marco|    Brasil|
|Helena|   30|  true|Helena|    Brasil|
+------+-----+------+------+----------+
'''
# tipo de join
join_type = "inner"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```

```
'''
join_type = "left_semi"

retorna todas as colunas do DataFrame / Dataset esquerdo e ignora todas as colunas do dataset direito.
Em outras palavras, esta junção retorna colunas do único conjunto de dados esquerdo para a correspondência de registros no conjunto de dados direito na expressão de junção, os registros não combinados na expressão de junção são ignorados nos conjuntos de dados esquerdo e direito.

+------+-----+------+
|  nome|idade|status|
+------+-----+------+
| Vasco|   62|  true|
| Marco|   33|  true|
|Helena|   30|  true|
+------+-----+------+
'''
# tipo de join
join_type = "left_semi"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```

```
'''
join_type = "right_outer"

é o oposto de left_join, aqui ele retorna todas as linhas do DataFrame/Dataset direito, independentemente da matemática encontrada no dataset esquerdo, quando a expressão de junção não corresponde, ele atribui null para esse registro e descarta os registros da esquerda onde não correspondem encontrado.

+------+-----+------+-------+----------+
|  nome|idade|status|   nome|PaisOrigem|
+------+-----+------+-------+----------+
| Vasco|   62|  true|  Vasco|  Portugal|
| Marco|   33|  true|  Marco|    Brasil|
|  null| null|  null|Gabriel|    Brasil|
|Helena|   30|  true| Helena|    Brasil|
|  null| null|  null|  Medhi|   Marocco|
+------+-----+------+-------+----------+
'''
# tipo de join
join_type = "right_outer"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```

```
'''
val join_type = "left_outer"

retorna todas as linhas do DataFrame / Dataset esquerdo, independentemente da correspondência encontrada no dataset direito, quando a expressão de junção não corresponde, ele atribui nulo para esse registro e elimina os registros da direita onde a correspondência não foi encontrada.

+--------+-----+------+------+----------+
|    nome|idade|status|  nome|PaisOrigem|
+--------+-----+------+------+----------+
| Antonio|   33|  true|  null|      null|
|   Vasco|   62|  true| Vasco|  Portugal|
|   Marco|   33|  true| Marco|    Brasil|
| Pereira|   33|  true|  null|      null|
|  Carlos|   28|  true|  null|      null|
|Fernando|   35|  true|  null|      null|
| Candido|   75| false|  null|      null|
|  Helena|   30|  true|Helena|    Brasil|
|    Lisa|   26|  true|  null|      null|
+--------+-----+------+------+----------+
'''
# tipo de join
join_type = "left_outer"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```

```
'''
join_type = "full_outer"

retorna todas as linhas de Spark DataFrame/Datasets, onde a expressão de junção não corresponde, ela retorna nulo nas respectivas colunas de registro.

+--------+-----+------+-------+----------+
|    nome|idade|status|   nome|PaisOrigem|
+--------+-----+------+-------+----------+
| Antonio|   33|  true|   null|      null|
|   Vasco|   62|  true|  Vasco|  Portugal|
|   Marco|   33|  true|  Marco|    Brasil|
| Pereira|   33|  true|   null|      null|
|  Carlos|   28|  true|   null|      null|
|    null| null|  null|Gabriel|    Brasil|
|Fernando|   35|  true|   null|      null|
| Candido|   75| false|   null|      null|
|  Helena|   30|  true| Helena|    Brasil|
|    Lisa|   26|  true|   null|      null|
|    null| null|  null|  Medhi|   Marocco|
+--------+-----+------+-------+----------+
'''
# tipo de join
join_type = "full_outer"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```

```
'''
join_type = "left_anti"

faz exatamente o oposto da left_semi, retorna apenas colunas do DataFrame/Dataset esquerdo para registros não correspondentes.

+--------+-----+------+
|    nome|idade|status|
+--------+-----+------+
| Antonio|   33|  true|
| Pereira|   33|  true|
|  Carlos|   28|  true|
|Fernando|   35|  true|
| Candido|   75| false|
|    Lisa|   26|  true|
+--------+-----+------+
'''
# tipo de join
join_type = "left_anti"
# condição
join_condition = dff1.nome == dff2.nome

df3 = dff1.join(dff2, join_condition, join_type)
df3.show()
```
